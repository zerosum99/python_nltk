```python
import nltk
```

## 코퍼스 내의 brown을 확인

    실제 코퍼스에 등록되어 있는 자료에 대한 세부 문서들을 확인하기

```python
from nltk.corpus import brown

print(brown.fileids())
```

## 코퍼스 내의 특정 문서에 대한 정보를 확인

    특정 문서를 조회하기 위기 raw 메소드를 사용하기 

```python
brown.raw('ca01')[:50]
```

## 코퍼스 내의 특정 문서에 대한 단어만 추출 

    특정 문서에 대해 단어별로 토근화해서 조회하기
    

```python
from nltk.corpus import brown

for i in dir(brown) :
    if not i.startswith("_") :
        print(i)
```

##  실제 있는 위치를 확인하기


```python
print(brown.root)
print(brown.abspath('ca01'))
print(brown.abspaths()[:3])
```

##  기본적인 설명 확인하기

```python
brown.readme()
```

## 문장에 대해 분리하기 

```python
brown.sents('ca02')
```

## 문장의 단어별로 실제 tag인 품사 표시



```python
brown.tagged_sents('ca02')
```

## 단어별로 분리하기

```python
print(brown.paras())
```

```python
from nltk.corpus import brown
brown.words('ca01')
```

## 단어별로 실제 tag인 품사 표시


```python
brown.tagged_words('ca01')
```

```python
brown.tagged_words()[0:10]
```
