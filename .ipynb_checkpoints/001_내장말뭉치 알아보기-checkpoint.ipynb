## 주제별 범주 알아보기

```python
reuters.categories()
```

```python
count = 0
for w in reuters.words(categories=['bop','cocoa']) :
    count += 1
    print(w + "   ", end="")
    
    if (w is '.') :
        print()
        
    if count > 100 :
        break
```
